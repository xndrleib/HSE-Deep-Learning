If you've skipped the url above out of sloth, here's the game of life:
* You have a 2D grid of cells, where each cell is "alive"(1) or "dead"(0)
* Any living cell that has 2 or 3 neighbors survives, else it dies [0,1 or 4+ neighbors]
* Any cell with exactly 3 neighbors becomes alive (if it was dead)

For this task, you are given a reference numpy implementation that you must convert to pytorch.
numpy code inspired by: https://github.com/rougier/numpy-100


__Note:__ You can find convolution in `torch.nn.functional.conv2d(Z,filters)`. Note that it has a different input format. 


```python
from scipy.signal import correlate2d as conv2d


def np_update(Z):
    # Count neighbours with convolution
    filters = np.array([[1, 1, 1],
                        [1, 0, 1],
                        [1, 1, 1]])

    N = conv2d(Z, filters, mode='same')

    # Apply rules
    birth = (N == 3) & (Z == 0)
    survive = ((N == 2) | (N == 3)) & (Z == 1)

    Z[:] = birth | survive
    return Z
```

```python
def torch_update(Z):
    """
    Implement an update function that does to Z exactly the same as np_update.
    :param Z: torch.FloatTensor of shape [height,width] containing 0s(dead) an 1s(alive)
    :returns: torch.FloatTensor Z after updates.
    
    You can opt to create new tensor or change Z inplace.
    """

    filters = torch.tensor([[1, 1, 1],
                            [1, 0, 1],
                            [1, 1, 1]], dtype=torch.float32)

    N = torch.nn.functional.conv2d(Z[None, None, ...], filters[None, None, ...], padding=1)

    # Apply rules
    birth = (N == 3) & (Z[None, None, ...] == 0)
    survive = ((N == 2) | (N == 3)) & (Z[None, None, ...] == 1)

    Z[:] = birth | survive
    return Z
```

```python
#initial frame
Z_numpy = np.random.choice([0, 1], p=(0.5, 0.5), size=(100, 100))
Z = torch.from_numpy(Z_numpy).type(torch.FloatTensor)

#your debug polygon :)
Z_new = torch_update(Z.clone())

#tests
Z_reference = np_update(Z_numpy.copy())
assert np.all(
    Z_new.numpy() == Z_reference), "your pytorch implementation doesn't match np_update. Look into Z and np_update(ZZ) to investigate."
print("Well done!")
```

```python
from notmnist import load_notmnist

X_train, y_train, X_test, y_test = load_notmnist(letters='ABCDEFGHIJ')
X_train, X_test = X_train.reshape([-1, 784]), X_test.reshape([-1, 784])
```

```python
%matplotlib inline
plt.figure(figsize=[12, 4])
for i in range(20):
    plt.subplot(2, 10, i + 1)
    plt.imshow(X_train[i].reshape([28, 28]))
    plt.title(str(y_train[i]))
```

```python
from torch.distributions.uniform import Uniform
import math


def xavier_initialization(n, m):
    r = 1 / math.sqrt(n)
    return Uniform(-r, r).sample((n, m)).requires_grad_(True)
```

```python
n_epoch = 100
lr = 1

history = []
train_log = []
val_log = []

w1 = xavier_initialization(784, 100)
b1 = torch.zeros(100, requires_grad=True)

w2 = xavier_initialization(100, 10)
b2 = torch.zeros(10, requires_grad=True)


def model(x_ar):
    # Linear
    l1_out = x_ar @ w1 + b1
    # Sigmoid
    l2_out = F.sigmoid(l1_out)
    # Linear
    l3_out = l2_out @ w2 + b2
    #Softmax
    return F.log_softmax(l3_out, dim=1)


def iterate_minibatches(inputs, targets, batchsize, shuffle=False):
    assert len(inputs) == len(targets)
    if shuffle:
        indices = np.random.permutation(len(inputs))
    for start_idx in range(0, len(inputs) - batchsize + 1, batchsize):
        if shuffle:
            excerpt = indices[start_idx:start_idx + batchsize]
        else:
            excerpt = slice(start_idx, start_idx + batchsize)
        yield inputs[excerpt], targets[excerpt]


for epoch in range(n_epoch):
    if epoch == 0:
        with torch.no_grad():
            predicted_y_train = np.argmax(model(torch.tensor(X_train, dtype=torch.float32)).numpy(), axis=1)
            predicted_y_test = np.argmax(model(torch.tensor(X_test, dtype=torch.float32)).numpy(), axis=1)

        train_log.append(np.mean(predicted_y_train == y_train))
        val_log.append(np.mean(predicted_y_test == y_test))

    for x_batch, y_batch in iterate_minibatches(X_train, y_train, batchsize=256, shuffle=True):
        x_batch = torch.tensor(x_batch, dtype=torch.float32)
        y_batch = torch.tensor(y_batch, dtype=torch.float32)

        y_pred = model(x_batch)

        loss = F.nll_loss(y_pred, y_batch.type(torch.LongTensor))
        loss.backward()

        with torch.no_grad():
            w1 -= lr * w1.grad
            b1 -= lr * b1.grad
            w2 -= lr * w2.grad
            b2 -= lr * b2.grad

            w1.grad.zero_()
            b1.grad.zero_()
            w2.grad.zero_()
            b2.grad.zero_()

    history.append(loss.data.numpy())

    if epoch % 10 == 0:
        print("epoch #%i | mean loss = %.3f" % (epoch, np.mean(history[-10:])))
    elif epoch == n_epoch - 1:
        print("epoch #%i | mean loss = %.3f" % (epoch, np.mean(history[-10:])))

    with torch.no_grad():
        predicted_y_train = np.argmax(model(torch.tensor(X_train, dtype=torch.float32)).numpy(), axis=1)
        predicted_y_test = np.argmax(model(torch.tensor(X_test, dtype=torch.float32)).numpy(), axis=1)

    train_log.append(np.mean(predicted_y_train == y_train))
    val_log.append(np.mean(predicted_y_test == y_test))
```

```python
fig = plt.figure(figsize=(10, 10))
fig.set_facecolor('xkcd:white')

plt.plot(train_log, label='train accuracy')
plt.plot(val_log, label='val accuracy')
plt.xlabel('Epoch')
plt.legend(loc='best')
plt.grid()
plt.show()
```

### Evaluation

Let's see how our model performs on test data

```python
with torch.no_grad():
    predicted_y_test = model(torch.tensor(X_test, dtype=torch.float32)).numpy()

predicted_y_test = np.argmax(predicted_y_test, axis=1)

assert isinstance(predicted_y_test, np.ndarray), "please return np array, not %s" % type(
    predicted_y_test)
assert predicted_y_test.shape == y_test.shape, "please predict one class for each test sample"
assert np.all(np.in1d(predicted_y_test, y_test)), "please predict class indexes"

accuracy = np.mean(predicted_y_test == y_test)
print("Test accuracy: %.3f" % accuracy)
```
